```python
alive=titanic['alive'].value_counts()
```

```python
titanic['alive'].unique()
```

```python
plt.pie(alive,labels=titanic['alive'].unique(),autopct='1%.1f%%',colors=['#db291d','#24c219'])
plt.show()
```

```python
sns.catplot(data=titanic,x='survived',y='age',kind='swarm',col='sex',sharey=False)
```

* 1 indicate survived and 0 indicate the died
- this show the survived and died people in the ship by their age 
- female are more survived than the male
- in between the age of 15 to 50 are the most died male
- in between 20 to 35 aged male are survived the most
- in between the age of 10 to 50 aged females more survived
